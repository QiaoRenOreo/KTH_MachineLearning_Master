```python

import numpy as np

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM


data = [[i for i in range(100)]]
data = np.array(data, dtype=float)
# print (data.shape)
# print (data[0][1])


target = [[i for i in range(1,101)]]
target = np.array(target, dtype=float)

data = data.reshape((1, 1, 100)) 
target = target.reshape((1, 1, 100)) 
print (data.shape)
print (target.shape)
# print (target)
```

```python
x_test=[i for i in range(100,200)]
x_test=np.array(x_test).reshape((1,1,100))
y_test=[i for i in range(101,201)]
y_test=np.array(y_test).reshape(1,1,100)
print(data.shape)
print(target.shape)
print(x_test.shape)
print(y_test.shape)
```

```python



model = Sequential()  
model.add(LSTM(10, input_shape=(1, 100),return_sequences=True))
model.add(Dense(100))
model.compile(loss='mean_absolute_error', optimizer='adam',metrics=['accuracy'])
history=model.fit(data, target, nb_epoch=10000, batch_size=1, verbose=2,validation_data=(x_test, y_test))




```

```python
model.summary()
```

```python


results = model.predict(x_test)
print (results.shape)
print (y_test.shape)
plt.scatter (range(100), results, c='r')
plt.scatter (range(100), y_test, c='g')
plt.show()
```

```python
plt.plot (history.history['loss'])
plt.show ()
```
